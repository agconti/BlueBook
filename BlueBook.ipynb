##Kaggle Competition | Blue Book for Bulldozers Quantitative Model
This IPython Notebook contains a quantitative pricing model created for [Fast Iron](http://www.fastiron.com/) in the Kaggle competition 'Blue Book for Bulldozers'. The model predicts the sale price of a particular piece of heavy equipment so that Fast Iron can create a 'Blue Book' to enable customers to valuate their heavy equipment fleet at auction.  Here I use Python as a medium to apply supervised and unsupervised machine learning techniques to explain 88.90% of the variance observed in the training set and score an RMSLE of 0.745 when predicting auction sale prices on the test set. In this competition 590 data scientists created models to compete for a $10,000 dollar award for the team or individual with the most accurate model.
```python
# New data
GDP.DATE = GDP.DATE.apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d")) # Coverts string dates into python datetime objects
inflation.DATE = inflation.DATE.apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
Industrial_production.DATE = Industrial_production.DATE.apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d"))
cat.Date = cat.Date.apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y"))
# Our original data
df.saledate = df.saledate.apply(lambda x: datetime.datetime.strptime(str(x), '%m/%d/%Y %H:%M')) 
```

### Merge data on the machine appendix to correct for data quality . 

```python
df = pd.merge(df, machine_appendix, left_on='MachineID', right_on='MachineID', how='right', suffixes=('_orginal','_machine_appendix'))
```

### Fill data forward to make a complete data calender as auctions happend 24/7 and not just on trading or business days.

```python
# Prepare data for forward filling
GDP = GDP.set_index('DATE')
inflation = inflation.set_index('DATE')
Industrial_production = Industrial_production.set_index('DATE')
cat = cat.set_index('Date')

# Fill data forward
GDP = GDP.resample('D', fill_method='ffill')
inflation = inflation.resample('D', fill_method='ffill')
Industrial_production = Industrial_production.resample('D', fill_method='ffill')
cat = cat.resample('D', fill_method='ffill')
```

### Lag our new variales and then merge them to our dataset:

```python
df = pd.merge(GDP.shift(90, freq='D'), df, left_index=True, right_on='saledate', how='right')
df = pd.merge(inflation.shift(90, freq='D'), df, left_index=True, right_on='saledate', how='right')
df = pd.merge(Industrial_production.shift(90, freq='D'), df, left_index=True, right_on='saledate', how='right')
df = pd.merge(cat.shift(90, freq='D'), df, left_index=True, right_on='saledate', how='right')
```

###Add some helpful functions to clean the dozer manufacturing date data and prepare it for our analysis. 


```python
def dozeryearclean(s):
    '''
    changes the bulldozers that were made in 1000 A.D. to a NaN so that they can be removed
    '''
    if s == 1000:
        s = np.nan
    return s

def MfgClean(s):
    '''
    changes non years in the Manufacturing year column to the correct value of NaN
    '''
    if isinstance(s, str):
        s = np.nan
    return s 
```

```python
df.YearMade = df.YearMade.apply(dozeryearclean) 
df.MfgYear = df.MfgYear.apply(MfgClean)          
df['Quarter'] = df.saledate.apply(agc.quater_maker)# adds a column for that specifices the quater each machine was sold in.

df = df.dropna() # Drops Null Values from the dataframe.
```

###Let's start our analysis. Here are my Regressions:
First, let’s start with a basic ordinary least squares regression and then refine our model. Let’s use GDP to compensate for the changes in the prices, or supply / demand changes, of the heavy machinery.

###Regression 1: Basic regression using GDP

The three cells below show the process of creating our model, training it on the data, and examining the model's performance. 

First, define our formula for our OLS regression in the first cell. In the next cell we create a regression friendly dataframe that sets up boolean values and lets our regression model know the types of inputs we're giving it. The model is then instantiated and fitted before a summary of the model's performance is printed. In the last cell we graphically compare the predictions of our model to the actual values we are trying to predict, as well as the residual errors from our model to check for any structure we may have missed. Below is the process carried out for our 'Basic regression using GDP':


```python
formula = 'SalePrice ~ C(YearMade) + C(ModelID_machine_appendix) + C(UsageBand) + C(auctioneerID) + GDP'
```

```python
y, x = dmatrices(formula, data=df, return_type='dataframe')# Prepare dataframe for regression
mod = sm.OLS(y, x)   # Describe model, so it can be fitted  
res1 = mod.fit()     # Fit model

results={}           # create a bin to hold the results of all of our models
results['Basic_regression_using_GDP']=[res1, formula] # save results for later
res1.summary()       # Summarize model
```

```python
# Plot Predictions vs Actual
plt.figure(figsize(18,6), dpi=1600);
plt.subplot(121, axisbg="#DBDBDB")
ypred = res1.predict(x) # use our model to make predictions on the test set
plt.loglog(x.index, ypred, 'bo', x.index, y, 'mo', alpha=.25);
plt.grid(color='white', linestyle='dashed')
plt.title('Basic_regression_using_GDP\nOLS prediction, Blue: \nFitted/predicted values: Red');
# Residuals
plt.subplot(122, axisbg="#DBDBDB")
plt.plot(res1.resid, 'r-')
plt.grid(color='white', linestyle='dashed')
plt.title('Basic_regression_using_GDP');
```

###Regression 2: My submitted model

After taking a look at `Basic_regression_using_GDP`'s ‘Predicted vs. Actual Values’ plot, there seemed to be some structure at the larger X values that our model missed.  To advance my model past the first basic regression and to better explain that missed structure, I changed the way I used `SalePrice` and removed the `GDP` feature. 

I transformed the `SalePrice` variable by taking the natural log of its observations so that its predictions could be better expressed by our linear model. Because GDP is an aggregate of all products produced in the US, from Battleships to Starbucks lattes, the noise created from production of all other goods made GDP a poor proxy for predicting the specific production, supply / demand, of heavy machinery in my own opinion . Increased governmental spending and expansionary fiscal policies after the 2008 financial crisis meant that the percentage of public spending to private spending was greater for the majority of the observations, and so may have decreased the explanatory power GDP for bulldozer production further. For these reasons I removed the `GDP` feature from our model and opted for a different, more comprehensive and targeted approach.

To do so, I added several additional features to more precisely capture the variance I was trying to explain with the `GDP` feature. These features were:

<ul> 
    <li>Inflation Index</li>
    <li>Industrial Production Index</li>
    <li>log of Caterpillar's common stock's adjusted closing price</li>
    <li>ModelID from the merged machine_appendix</li>
    <li>the quarter the machine was sold in</li>
</ul>

Because of Caterpillar's dominating and wide representing position in the heavy machinery industry based on market capitalization and revenue, I included the stock price of the Caterpillar as a proxy for the private expectations of growth for heavy machinery industry. I included an Industrial Production index from the FRED to directly target the previous quarters output, as the variable was lagged 90 days. I also adjusted for nominal pricing changes between the sale prices of the bulldozers by including a CPI index for the timeframe sampled. To further adjust for the seasonality in the supply and demand of the bulldozers, I included a Quarter variable that could adjust the sale price for the season the equipment was sold in.


```python
formula = 'np.log(SalePrice) ~ Inflation_Index + Industrial_production_index + np.log(cat_adj_close) + C(YearMade) + C(ModelID_machine_appendix) + C(Quarter) + C(auctioneerID) + C(UsageBand)'
```

```python
y, x = dmatrices(formula, data=df, return_type='dataframe')
mod = sm.OLS(y, x)   # Describe model, so it can be fitted C(UsageBand)
res2 = mod.fit()     # Fit model
results['My_final_regression']=[res2, formula]
res2.summary()       # Summarize model 
```

```python
plt.figure(figsize(18,6), dpi=1600)

# Predictions vs. actual
plt.subplot(121, axisbg="#DBDBDB")
ypred = res2.predict(x)
plt.loglog(x.index, ypred, 'bo', x.index, y, 'mo', alpha=.25);
plt.grid(color='white', linestyle='dashed')
plt.title('My_final_regression\nOLS prediction, Blue: \nFitted/predicted values: Red');

# Residuals
plt.subplot(122, axisbg="#DBDBDB")
plt.plot(res1.resid, 'r-')
plt.grid(color='white', linestyle='dashed')
plt.title('My_final_regression Residuals');
```

Our transformation of the dependent variable and the addition of our new features seem to have explained the structure that we initially missed with our first regression. The resulting Predicted vs. Actual values plot for `My_final_regression`, now looks appropriate for a linear model. But, it’s time for a gut check.  After reviewing our formula and the results of our model, I suspected over fitting from the `ModelID` variable because there were so many unique occurrences of each machine’s model, many models found in our training set do not appear in our test set, and many models in the test set do not appear in our training set. After coming to this conclusion, I adjusted my previous regression by replacing `ModelID` with two proxies, `fiBaseModel_machine_appendix` and `fiProductClassDesc_machine_appendix`, so that it might perform better on the test set.

###Regresion 3: A variation of the previous regression that exchanges the ModelID feature for fiBaseModel_machine_appendix and fiProductClassDesc_machine_appendix to adjust for possible over fitting.


```python
formula = 'np.log(SalePrice) ~ fiBaseModel_machine_appendix + C(fiProductClassDesc_machine_appendix) + Inflation_Index + Industrial_production_index + np.log(cat_adj_close) + C(YearMade) + C(Quarter) + C(auctioneerID) + C(UsageBand)'
```

```python
y, x = dmatrices(formula, data= df, return_type='dataframe')
mod = sm.OLS(y, x)    # Describe model, so it can be fitted  fredb
res3 = mod.fit()      # Fit model
results['My_regression_corrected_for_overfitting']=[res3, formula]
res3.summary()        # Summarize model 
```

```python
plt.figure(figsize(18,6))
ypred = res3.predict(x)

plt.subplot(121, axisbg="#DBDBDB")
plt.loglog(x.index, ypred, 'bo', x.index, y, 'mo', alpha=.25);
plt.grid(color='white', linestyle='dashed')
plt.title('My_regression_corrected_for_overfitting\nOLS prediction, Blue: \nFitted/predicted values: Red');

# Residuals
plt.subplot(122, axisbg="#DBDBDB")
plt.plot(res1.resid, 'r-')
plt.grid(color='white', linestyle='dashed')
plt.title('My_regression_corrected_for_overfitting');
```

###Now that we have our models, lets generate our predictions on the test and validation sets and see how they perform.

### Merge data on the machine appendix to correct for data quality .

###Predict and Output our models:

```python
for i in results:
    prog = 0                                       # variable to keep track of the progress of the output.
    check_points = 6                               # total number of progress points per iteration.
    model_params = DataFrame(results[i][0].params) # extract the parameters of the model
    formula = results[i][1]                        # extract the formula of the model
    print "\nProcessing: " + i
    prog += 1
    agc.progress(prog, check_points)               # a simple textual progress bar
    
    # Create reg friendly test dataframe
    yt, xt = dmatrices(formula, data=test_data, return_type='dataframe')
    prog += 1
    agc.progress(prog, check_points)
    
    # remove extraneous features for efficiency 
    to_drop = list()
    to_drop[:] = []                    # Empty list, in case cells are executed out of order
    for c in xt.columns:               # find features to drop
        if c not in model_params.index:
            to_drop.append(c)
    xt = xt.drop(to_drop, axis=1)      # drop features
    
    # find parameters not contained in the test set and remove them for efficiency.
    to_drop[:] = [] # Empty list
    for c in model_params.index : 
        if c not in xt.columns:
            to_drop.append(c)
    model_params = model_params.drop(to_drop)
    prog += 1
    agc.progress(prog, check_points)
    
    # Convert to NumPy arrays for performance
    model_params = np.asarray(model_params)
    yt = np.asarray(yt)
    yt = yt.ravel() 
    xt = np.asarray(xt)
    prog += 1
    agc.progress(prog, check_points)
    
    # Use our models to create predictions
    row, col = xt.shape
    model_params = model_params.ravel() # flatten array
    model_array = []
    
    for _ in xrange(row):
            model_array.append(model_params)
    model_array = np.asarray(model_array)
    
    # Multiply matrix together 
    predictions = np.multiply(xt, model_array)
    predictions = np.sum(predictions, axis = 1) # calculate sale price
    predictions = np.exp(predictions)           # transform sale price to a linear scale
    prog += 1
    agc.progress(prog, check_points)
    
    # Output Results so our performance on the test set can be scored by Kaggle
    predictions = Series(predictions, name=['SalePrice'])# Put the NumPy array into a Pandas Series for effecient output to Kaggle
    predictions.to_csv(i + ".csv", na_rep=0, float_format='%.3f', index_label=False) # output predictions
    prog += 1
    agc.progress(prog, check_points)
    
```

###Score Our Results:

```python
# Load in solved Cross Validation dataset for scoring
valid_solution = pd.read_csv('ValidSolution.csv')
valid = pd.read_csv('Valid.csv')
valid['SalePrice'] = np.random.randn() # add the missing independent variable
```

```python
for i in results:
    prog = 0                         # variable to keep track of the progress of the output.
    check_points = 8                 # total number of progress points per iteration.
    model_params = DataFrame(results[i][0].params)
    formula = results[i][1]
    print "\nProcessing: " + i
    prog += 1
    agc.progress(prog, check_points) # a simple textual progress bar
   
    
    yt, xt = dmatrices(formula, data=valid.fillna(value=0), return_type='dataframe') 
    prog += 1
    agc.progress(prog, check_points)

    
    # remove extraneous features for efficeny 
    to_drop[:] = []
    to_drop = list()
    for c in xt.columns:
        if c not in model_params.index:
            to_drop.append(c)
    xt = xt.drop(to_drop, axis=1)
    
    #find parameters not contained in the test set and remove them for efficiency.
    to_drop[:] = [] # Empty list
    for c in model_params.index : 
        if c not in xt.columns:
            to_drop.append(c)
    model_params = model_params.drop(to_drop)
    prog += 1
    agc.progress(prog, check_points)
    
    # Convert to NumPy arrays for performance
    model_params = np.asarray(model_params)
    yt = np.asarray(yt)
    yt = yt.ravel()
    xt = np.asarray(xt)
    prog += 1
    agc.progress(prog, check_points)
    
    # Generate predictions from our training set
    row, col = xt.shape
    model_params = model_params.ravel() # flatten array
    model_array = []
    for _ in xrange(row):
            model_array.append(model_params)
    model_array = np.asarray(model_array)
    prog += 1
    agc.progress(prog, check_points)
   
    # Multiply matrix together 
    predictions = np.multiply(xt, model_array)
    predictions = np.sum(predictions, axis = 1)
    predictions = np.exp(predictions)    
    prog += 1
    agc.progress(prog, check_points)
    
    # Score the results based on our validation set
    actual = np.asarray(valid_solution.SalePrice)
    rmsle = np.sqrt(np.mean(np.power(np.log(actual + 1) - np.log(predictions + 1), 2)))
    prog += 1
    agc.progress(prog, check_points)
    
    rmse = np.sqrt(np.mean(np.power((actual - predictions), 2)))
    prog += 1
    agc.progress(prog, check_points)
    
    print " %s scored an RMSE of : %.4f and an RMSLE of : %.4f\n" %(i, rmse, rmsle)
```

###The results and post mortem thoughts:

`My_final_regression` and `My_regression_corrected_for_overfitting` scored a RMSLE of 1.17307 and 0.75255 respectively on the test set. I selected my `My_final_regression`, not corrected for over fitting, for submission and finished in 108th place. The winning model conducted by the team Leustagos & Titericz scored an RMSLE of 0.22773 by using an ensemble of GLM models. 

What could I have done to improve my own model?

If I had used my second model, which corrected for the over fitting effect of the `ModelID` variable, I would have placed 104th. But what could I have done to get in the top models? A random forest would have done the trick; it’s shown below. It would have scored an RMSLE of 0.26704, only 0.039 RMSLE greater than the winning model, and I would have placed 60th, and in the top 10%, if submitted. The difference between the first place model and the 60th is small, and could have been overcame by training the model on only the previous 6 months of observations for a given prediction as team Leustagos & Titericz did. 


```python
from sklearn.ensemble import RandomForestRegressor
prog = 0 
total_tasks = 7

# Remove duplicates from merge
valid = valid.drop(['fiModelDesc_orginal', 'fiBaseModel_orginal', 'fiSecondaryDesc_orginal', 'fiModelSeries_orginal', 'fiModelDescriptor_orginal', 
           'ProductSize', 'fiProductClassDesc_orginal','ProductGroup_orginal', 'ProductGroupDesc_orginal'], axis=1)
prog += 1
agc.progress(prog, total_tasks)

# Create Random Forest Formula
formula = 'SalePrice ~'
for i, val in enumerate(valid.columns):
    if i != 0 and val != 'SalePrice':
        formula += ' + ' + val 

prog += 1
agc.progress(prog, total_tasks)

# Create friendly dataframe to fit model to 
y, x = dmatrices(formula, data=valid.fillna(value=(-1)), return_type='dataframe')
prog += 1
agc.progress(prog, total_tasks)

# instantiate model
rf = RandomForestRegressor(n_estimators=50, n_jobs=(1))
prog += 1
agc.progress(prog, total_tasks)

# fit model
results = rf.fit(x, y)
prog += 1
agc.progress(prog, total_tasks)


# Predict Model
predictions = results.predict(x)
prog += 1
agc.progress(prog, total_tasks)

# Score the results based on our validation set
actual = np.asarray(valid_solution.SalePrice)
rmsle = np.sqrt(np.mean(np.power(np.log(actual + 1) - np.log(predictions + 1), 2)))
rmse = np.sqrt(np.mean(np.power((actual - predictions), 2)))
prog += 1
agc.progress(prog, total_tasks)
print "The random forest scored an RMSE of : %.4f and an RMSLE of : %.4f\n" %(rmse, rmsle)
```
